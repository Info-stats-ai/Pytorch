```python
import time
```

```python
size =10000
```

```python
matrix_cpu1 = torch.rand(size,size)
matrix_cpu2 = torch.rand(size,size)
```

```python
# measure the time pon cpu of matrix multiplication
start = time.time()
result = matrix_cpu1 @ matrix_cpu2
end = time.time()
print(end-start)
```

```python
# moving matrices to gpu
matrix_gpu1 = matrix_cpu1.to(device)
matrix_gpu2 = matrix_cpu2.to(device)
start = time.time()
result = matrix_gpu1 @ matrix_gpu2
end = time.time()
print(end-start)
```
